```python
#CARI DATA MAKSIMAL CARA 1
maxAll = list(map(lambda a: dfC[dfC[a] == dfC[a].max()],header[0:6]))
maxAll
```

```python
#Data Frame tiap tahun yang lebih dari nol dengan function
def cleanUp(thn):
    tidakNol = dfC[dfC[thn]>0]
    return tidakNol[tidakNol[thn]==tidakNol[thn].min()]
# cleanUp(2003)
# cleanUp(2005)
# cleanUp(2008)
# cleanUp(2011)
# cleanUp(2014)
```

```python
#Cari Data Minimal Desa di Provinsi mana tiap tahunnya
minAll = list(map(cleanUp,header[0:6]))
minAll
```
