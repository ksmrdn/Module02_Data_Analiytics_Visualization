```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dfAsean = pd.read_csv('asean.csv')
dfAsean
```

```python
dfAsean = dfAsean.sort_values(by=['Name'])
dfAsean
```

```python
#BAR CHART POPULATION
plt.style.use('ggplot')
plt.figure('Population ASEAN',figsize=(18,10))
asGraph = plt.bar(dfAsean['Name'],dfAsean['Population'],color=['black', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'lightgreen', 'cyan', 'lightblue', 'blue'])
plt.xlabel('Negara', fontsize=14)
plt.ylabel('Population(x100 million)', fontsize=14)
plt.title('ASEAN Countries Population', fontsize=16)

for i in asGraph:
    print(i)
    height = i.get_height()
    print(height)
    print(i.get_x())
    plt.text(i.get_x() + i.get_width()/2, 
             (1.01*height),
             height,
             ha="center", va='bottom')
             
```

```python
#PIE CHART POPULATION
labels = dfAsean['Name']
sizes = dfAsean['Population']
explode = (0,0,0,0.1,0,0,0,0.025,0,0,0.05)
plt.figure(figsize=(15,8))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0, radius=2, colors=['black', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'lightgreen', 'cyan', 'lightblue', 'blue'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.suptitle('ASEAN Population Percentage')
plt.show()
```

```python
#BAR CHART GNP
plt.style.use('ggplot')
plt.figure('GNP ASEAN',figsize=(18,10))
asGraph = plt.bar(dfAsean['Name'],dfAsean['GNP'],color=['black', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'lightgreen', 'cyan', 'lightblue', 'blue'])
plt.xlabel('Negara', fontsize=14)
plt.ylabel('GNP($ US)', fontsize=14)
plt.title('ASEAN Countries GNP', fontsize=16)

for i in asGraph:
    print(i)
    height = i.get_height()
    print(height)
    print(i.get_x())
    plt.text(i.get_x() + i.get_width()/2, 
             (1.01*height),
             height,
             ha="center", va='bottom')
```

```python
#PIE CHART AREAS
labels = dfAsean['Name']
sizes = dfAsean['SurfaceArea']
explode = (0,0,0,0.1,0,0,0,0.025,0,0,0.05)
plt.figure(figsize=(15,8))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0, radius=2, colors=['black', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'lightgreen', 'cyan', 'lightblue', 'blue'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.suptitle('ASEAN Surface Area')
plt.show()
```
