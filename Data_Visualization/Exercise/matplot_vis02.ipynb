```python
#Pie chart, where the slices will be ordered annd plotted ccw
labels = ['Soto','Satay','Kebab','Sandwich']
sizes = [20,10,33,37]
explode = (0.05,0,0.1,0.15) #only explode teh 2nd slice
# fig1, ax1 = plt.subplots()

#template,outer,inner
x, y, z = plt.pie(sizes, explode= explode, labels=labels, autopct ='%1.2d%%',textprops={'size':10,'color':'k'},
           shadow=True, startangle=60, radius=1.2)
plt.axis('equal') #Equal aspect ratio ensures that pie is drawn as a circle.
for i in z:
    i.set_color('w')
```
