```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
p = plt.subplot(111,projection='3d')
data = range(11)
x = np.array(data)

p.plot_wireframe(
    x,x,x.reshape(1,-1),color='r',linestyle=':')
p.scatter(x,x,x,marker='*',color='y',s=150)
p.set_xlabel('X')
p.set_ylabel('Y')
p.set_zlabel('Z')
```

```python
plt.style.use('seaborn')
fig = plt.figure()
p = plt.subplot(111,projection='3d')
# p.scatter(x,y,z)
x = np.arange(10)
y = np.arange(10)
z = np.zeros(10)
dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x,y,z,dx,dy,dz,color = 'red')
p.set_xlabel('X')
p.set_ylabel('Y')
p.set_zlabel('Z')
plt.show()
```
