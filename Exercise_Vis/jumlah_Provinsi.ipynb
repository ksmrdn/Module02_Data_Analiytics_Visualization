```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sb
```

```python
dfPdd = pd.read_excel('pdd_12.xlsx',
                     skiprows=3,
                     skipfooter=3,
                     na_values='-',
                     )
dfPdd = dfPdd.fillna(0)
```

```python
dfPdd = dfPdd.rename(columns={"Unnamed: 0":"Provinsi"})
dfPdd
```

```python
x = np.arange(len(dfPdd['Provinsi'])) 
x
```

```python
y  = np.arange(1)
y
```

```python
z = np.zeros(len(dfPdd['Provinsi']))
z
```

```python
dx = np.ones(len(dfPdd['Provinsi']))
dx
```

```python
dy = np.ones(len(dfPdd['Provinsi']))
dy
```

```python
dz = dfPdd[1971]
dz
```

```python
# plt.style.use('seaborn-pastel')
fig = plt.figure(figsize=(25,20))
p = plt.subplot(111,projection='3d')
p.set_xlabel('Provinsi')
p.set_ylabel('Tahun 1971',fontsize=20)
p.set_zlabel('Jumlah Penduduk (Jiwa)',fontsize=20)
p.set_xticks(np.arange(len(dfPdd['Provinsi']))) 
p.set_xticklabels(dfPdd['Provinsi'], rotation=90)
p.set_yticks(np.arange(len(dfPdd['Provinsi'])))
p.set_yticklabels(np.arange(len(dfPdd['Provinsi'])))
p.bar3d(x,y,z,dx,dy,dz,color=['#484c7f','#ac8daf','#ddb6c6','#f1d4d4','#8186d5','#c6cbef','#e3e7f1','#3fc5f0','#42dee1','#6decb9','#eef5b2','#e6f8f9','#b1e8ed','#edb5f5','#e86ed0','#e5dfdf','#6ba8a9','#ac8daf','#1d4d4f','#494ca2','#6d0c74','#7f78d2','#d2d0fe','#fdecff','#5026a7','#8d448b','#cc6a87','#380e7f','#6915cf','#d62196','#e497cd','#6807f9','#c299fc'])
plt.show()
```
