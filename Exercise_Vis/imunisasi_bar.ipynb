```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
btBCG = pd.read_csv('BCG.csv',
                   na_values='n.a')
btBCG
```

```python
btBCG = btBCG.interpolate().fillna({})
btBCG
```

```python
btDPT = pd.read_csv('DPT.csv',
                    na_values='n.a')
btDPT
```

```python
btDPT = btDPT.interpolate().fillna({})
btDPT
```

```python
btCampak = pd.read_csv('Campak.csv',
                       na_values='n.a')
btCampak
```

```python
btCampak = btCampak.interpolate().fillna({})
btCampak
```

```python
btPolio = pd.read_csv('Polio.csv',
                      na_values='n.a')
btPolio
```

```python
btPolio = btPolio.interpolate().fillna({})
btPolio
```

```python
plt.figure('Persentasi balita terimunisasi 1995-2017', figsize = (15,8))

plt.subplot(221)
plt.bar(btBCG['Tahun'], btBCG['% Balita yang pernah mendapat imunisasi BCG'], color = 'r')
plt.title('BCG')
plt.xticks(btBCG['Tahun'], rotation = 90)

plt.subplot(222)
plt.bar(btCampak['Tahun'], btCampak['% Balita yang pernah mendapat imunisasi Campak'], color = 'g')
plt.title('Campak')
plt.xticks(btCampak['Tahun'], rotation = 90)

plt.subplot(223)
plt.bar(btDPT['Tahun'], btDPT['% Balita yang pernah mendapat imunisasi DPT'], color = 'y')
plt.title('DPT')
plt.xticks(btDPT['Tahun'], rotation = 90)

plt.subplot(224)
plt.bar(btPolio['Tahun'], btPolio['% Balita yang pernah mendapat imunisasi Polio'], color = 'b')
plt.title('Polio')
plt.xticks(btPolio['Tahun'], rotation = 90)
plt.tight_layout()

plt.show()
```

```python
plt.figure('Persentasi balita tak terimunisasi 1995-2017', figsize = (15,8))

plt.subplot(221)
plt.bar(btBCG['Tahun'], 100 - btBCG['% Balita yang pernah mendapat imunisasi BCG'], color = 'r')
plt.title('BCG')
plt.xticks(btBCG['Tahun'], rotation = 90)

plt.subplot(222)
plt.bar(btCampak['Tahun'], 100 - btCampak['% Balita yang pernah mendapat imunisasi Campak'], color = 'g')
plt.title('Campak')
plt.xticks(btCampak['Tahun'], rotation = 90)

plt.subplot(223)
plt.bar(btDPT['Tahun'], 100 - btDPT['% Balita yang pernah mendapat imunisasi DPT'], color = 'y')
plt.title('DPT')
plt.xticks(btDPT['Tahun'], rotation = 90)

plt.subplot(224)
plt.bar(btPolio['Tahun'], 100 - btPolio['% Balita yang pernah mendapat imunisasi Polio'], color = 'b')
plt.title('Polio')
plt.xticks(btPolio['Tahun'], rotation = 90)
plt.tight_layout()

plt.show()
```
