Buatlah sebuah file python yang mengakses data dari ke-4 file .csv tersebut, kemudian tampilkan data plot harga penutupan (close) harian seluruh data yang ada. Output yang diharapkan berupa grafik garis yang menampilkan harga penutupan (close) harian dari ke-4 perusahaan telco selama 3 bulan terakhir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#XL
dfE = pd.read_csv('EXCL.JK.csv',
                    parse_dates = ['Date']
                 )
dfE = dfE.set_index('Date')
```

```python
dfE = dfE.resample('D').sum()
dfE.iloc[:,:5] = dfE.iloc[:,:5].replace(0,np.NaN)
dfE['2019-03-18':'2019-03-31']
```

```python
dfE = dfE.fillna(method='ffill')
dfE['2019-03-18':'2019-03-31']
```

```python
#FREN
dfF = pd.read_csv('FREN.JK.csv',
                  parse_dates = ['Date']
                 )
dfF = dfF.set_index('Date')
```

```python
dfF = dfF.resample('D').sum()
dfF.iloc[:,:5] = dfF.iloc[:,:5].replace(0,np.NaN)
dfF['2019-03-18':'2019-03-31']
```

```python
dfF = dfF.fillna(method='ffill')
dfF['2019-03-18':'2019-03-31']
```

```python
#INDOSAT
dfI = pd.read_csv('ISAT.JK.csv',
                 parse_dates = ['Date'])
dfI = dfI.set_index('Date')
```

```python
dfI = dfI.resample('D').sum()
dfI.iloc[:,:5] = dfI.iloc[:,:5].replace(0,np.NaN)
dfI['2019-03-18':'2019-03-31']
```

```python
dfI = dfI.fillna(method='ffill')
dfI['2019-03-18':'2019-03-31']
```

```python
#TELKOM
dfT = pd.read_csv('TLKM.JK.csv',
                 parse_dates = ['Date'])
dfT = dfT.set_index('Date')
```

```python
dfT = dfT.resample('D').sum()
dfT.iloc[:,:5] = dfT.iloc[:,:5].replace(0,np.NaN)
dfT['2019-03-18':'2019-03-31']
```

```python
dfT = dfT.fillna(method='ffill')
dfT['2019-03-18':'2019-03-31']
```

```python
date1 = pd.date_range(start='2019-03-16', end='2019-06-19', freq='W-sat')
```

```python
plt.style.use('seaborn-pastel')
plt.figure('Saham',figsize=(16,8))
plt.grid('True')
plt.plot(dfE.index,dfE['Close'],label='PT XL Axiata Tbk')
plt.plot(dfF.index,dfF['Close'],label='PT Smartfren Telecom Tbk')
plt.plot(dfI.index,dfI['Close'],label='PT Indosat Tbk')
plt.plot(dfT.index,dfT['Close'],label='PT Telekomonikasi Indonesia')
plt.xlabel('Date',fontsize=14)
plt.ylabel('Rupiah (IDR)',fontsize=14)
plt.suptitle('Harga Historis Saham Provider Telco Indonesia',fontsize=16)
plt.xticks(date1,fontsize=10,rotation=45)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=4, mode="expand")
```

```python
dfE_Apr = dfE['2019-04-01':'2019-04-30']
dfF_Apr = dfF['2019-04-01':'2019-04-30']
dfI_Apr = dfI['2019-04-01':'2019-04-30']
dfT_Apr = dfT['2019-04-01':'2019-04-30']
```
