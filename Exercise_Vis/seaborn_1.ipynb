```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'x': np.arange(10),
    'y': np.arange(10),
    'z':['pria','wanita']*5
})
df
```

```python
plt.style.use('seaborn')
plt.plot(df.x, df.y)
```

```python
sns.set(style='whitegrid')
sns.relplot(data=df,x='x',y='y', hue='z',style='z')
```

```python
sns.barplot(data=df, x='x',y ='y', hue='z')
```

```python
sns.barplot(data=df, x='x',y ='y')
```
