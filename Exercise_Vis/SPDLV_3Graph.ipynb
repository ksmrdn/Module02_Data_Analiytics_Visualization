```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
#inisial array
a = np.array([[1,-2,1],
     [3,1,-2],
     [7,-6,-1]])
b = np.array([6,4,10])

#xyz values
pl = np.linalg.solve(a,b)
pl
```

```python
x_val = pl[0].round(0)
y_val = pl[1].round(1)
z_val = pl[2].round(2)
x_val, y_val, z_val
```

```python
#pers 1
x1 = np.array([6,0,0])
y1 = np.array([0,-3,0])
z1 = np.array([[0,0,6]])

#pers 2
x2 = np.array([1.33,0,0])
y2 = np.array([0,4,0])
z2 = np.array([[0,0,-2]])

#pers 3
x3 = np.array([1.43,0,0])
y3 = np.array([0,-1.67,0])
z3 = np.array([[0,0,-10]])
```

```python
plt.figure('SPDLV',figsize=(15,10))
```

```python
#plot first 3D
my_plot = plt.subplot(131,projection='3d')
my_plot.plot_wireframe(x1,y1,z1, facecolor='red', alpha=0.3)
my_plot.scatter(x1,y1,z1, color='blue', s=50)
my_plot.set_title('x - 2y + z = 6')
my_plot.set_xlabel('X Value')
my_plot.set_ylabel('Y Value')
my_plot.set_zlabel('Z Value')

#plot second 3D
my_plot = plt.subplot(132,projection='3d')
my_plot.plot_wireframe(x2,y2,z2, facecolor='yellow', alpha=0.3)
my_plot.scatter(x2,y2,z2, color='red', s=50)
my_plot.set_title('3x + y - 2z = 4')
my_plot.set_xlabel('X Value')
my_plot.set_ylabel('Y Value')
my_plot.set_zlabel('Z Value')

#plot third 3D
my_plot = plt.subplot(133,projection='3d')
my_plot.plot_wireframe(x3,y3,z3, facecolor='blue', alpha=0.3)
my_plot.scatter(x3,y3,z3, color='green', s=50)
my_plot.set_title('7x - 6y - z = 10')
my_plot.set_xlabel('X Value')
my_plot.set_ylabel('Y Value')
my_plot.set_zlabel('Z Value')


plt.subplots_adjust(left = 0, right = 3, wspace = 0.2)
```
