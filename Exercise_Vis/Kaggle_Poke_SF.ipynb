```python
import numpy as np
import pandas as pd
import folium as fl
import requests as req
```

```python
dfPoke = pd.read_csv('pokemon-spawns.csv')
dfPk = dfPoke.sample(1000,random_state=0)
dfUn = dfPk['name']
dfLat = dfPk['lat']
dfLan = dfPk['lng']
dfNum = dfPk['num']
```

```python
id_poke = dfNum.tolist()
un_poke = dfUn.tolist()
lat_poke = dfLat.tolist()
lan_poke = dfLan.tolist()

geo = list(zip(lat_poke,lan_poke))
geo
```

```python
png_poke = []
for i in range(len(dfPk.index.tolist())):
    input_id= dfPk.iloc[i,2]
    poke_url = f'https://pokeapi.co/api/v2/pokemon/{input_id}'
    poke_data = req.get(poke_url)
    data = poke_data.json()
    png_src = data['sprites']['front_default']
    png_poke.append(png_src)
```

```python
png_poke
```

```python
icon_poke = []
for j in range(len(un_poke)):
    glass = []
    glass.append(un_poke[j])
    glass.append(list(geo[j]))
    glass.append(png_poke[j])
    icon_poke.append(glass)
```

```python
icon_poke
```

```python
maps = fl.Map(
                location=[37.320157,-121.916599],
                zoom_start=10)
for k in range(len(icon_poke)):
    name = icon_poke[k][0]
    latlan = icon_poke[k][1]
    img_link = fl.features.CustomIcon(icon_poke[k][2],icon_size=(70,70))
    fl.Marker(
        location = latlan,
        tooltip = f"<b>{name}</b>",
        icon = img_link
    ).add_to(maps)
```
