```python
import pandas as pd
```

```python
dfSal = pd.read_csv('Salaries.csv')
dfSal
```

```python
dfYear = list(dfSal['Year'])
# dfYear
```

```python
# dfX = [2002,2003,2001,2000,2003,2004,2005,2006,2006]
listX = []
listY = [] 
for i in dfYear:
    if i not in listX:
        listY.append(i)
        listX.append(i)
        listY = []
print(listX)
```

```python
dfSal['Year'].unique()
```

```python
# dfX = [2002,2003,2001,2000,2003,2004,2005,2006,2006]
freq_table = {}
for i in dfYear:
    val = i
    if i not in freq_table:
        freq_table[val] = 1
    else:
        freq_table[val] += 1
print(freq_table)
```
