```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title')
# plt.show()
```

```python
#plt.subplot(nrows, ncols, plot_number)
plt.subplot(2,2,3)
plt.plot(x,y,'k-.')
plt.subplot(2,2,1)
plt.plot(x,y,'r-.')
plt.subplot(2,2,2)
plt.plot(y,x,'b.-')
```

```python
#create figure(empty canvas)
fig = plt.figure()

#add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height(range 0 to 1)

#Plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('set X')
axes.set_ylabel('set Y')
axes.set_title('Set Tittle')
```

```python
#create blank canvas
fig = plt.figure()
axes1 = fig.add_axes([0.0,0.0,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.1,0.4,0.4,0.3]) #insert axes

#larger figure axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('x_label1')
axes1.set_ylabel('y_label1')
axes1.set_title('Axes 1')

#Insert figure axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('x_label2')
axes2.set_ylabel('y_label2')
axes2.set_title('Axes 2')
```

```python
#Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x,y,'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
#Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=3)
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

#Display the figure object
fig
```

```python
#Figure size, aspect ratio and DPI
fig = plt.figure(figsize=(8,4), dpi=12)
```

```python
fig, axes = plt.subplots(figsize = (8,3))
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```
