```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[9,5,7,3,6],
    'z':[2,3,4,5,3]
})
plt.plot(df['x'],df['y'])
```

```python
plt.plot(df[['x','y']],df[['x','z']])
plt.grid('True')
plt.xticks(np.arange(0,10,step=1),rotation=0)
plt.xticks(np.arange(10), list('abcdefghijklmno'))
plt.yticks(np.arange(0,6,step=1),rotation=0)
```

```python
#STACK GRAPH

plt.bar(df.x,df.y, width=0.35, color=['pink','black'], edgecolor=['black','pink'],linewidth=2.5,yerr=0.2)

plt.bar(df.x,df.z, width=0.35, color=['k','pink'], edgecolor=['pink','k'],bottom=df.y,linewidth=2.5)

for i in df['y']:
    plt.text(
            df[df['y']==i]['x']-0.05,i+0.1,i) #x, #y, val
    
plt.title('This is chart', fontdict={'family':'impact','size':20,'color':'orange'})
```
