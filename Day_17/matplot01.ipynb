```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,3,5,8,9,1]
plt.plot(x,y)
```

```python
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,color='magenta',marker='1',linestyle=':')
plt.legend(['XY Line'])
plt.grid('True')
```

```python
z = [2,8,3,4,9,1]
plt.plot(x,y,'r-.',x,z,'b-.')
plt.grid('True')
plt.xlabel('X')
plt.ylabel('Y-Z')
plt.title('X Y Z')
plt.legend(['XY Line','XZ Line'],loc=8)
plt.subplots_adjust(left=0,right=1,top=0.75, bottom=0.1)
plt.text(0,0,'origin')
# plt.text(5,9,'peak')
# plt.savefig('XYZ.png')
for i,j in zip(x,y):
    plt.text(i,j,f'({i},{j})')
for k,l in zip(x,z):
    plt.text(k,l,f'({i},{j})')
# for i in z:
#     plt.text(z[list(z).index(i)]-.2,i+2,i)
plt.annotate(
        'Highest value!', xy=(5,9), xytext=(5.35,6.5),
    arrowprops = {
        'facecolor':'red','shrink':0.1,'width':3,'headwidth':7
    })
plt.show()
```
